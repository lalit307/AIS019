```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('datasets_46927_85203_data.csv')
```

```python
data.head()
```

```python
x=data.iloc[:,2:12]
y=data.iloc[:,1:2]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=202)
knn=KNeighborsRegressor(n_neighbors=3)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred.round(2)
```

**Mean square error and Root mean square eror**

```python
from sklearn.metrics import mean_squared_error
import math

for i in range(1,10):
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(x_train,y_train)
    y_pred = knn.predict(x_test)
    MSE = np.sqrt( mean_squared_error(y_test,y_pred))
    RMSE = math.sqrt(MSE)
    print('MSE when k is '+ str(i) + ' =',MSE)
    print('RMSE when k is '+ str(i) + ' =',RMSE)
```

```python
from sklearn.metrics import mean_absolute_error
import math

for i in range(1,10):
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(x_train,y_train)
    y_pred = knn.predict(x_test)
    MAE = np.sqrt( mean_absolute_error(y_test,y_pred))
    print('MAE when k is '+ str(i) + ' =',MAE)
```
